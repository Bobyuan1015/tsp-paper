```python
import pandas as pd

df = pd.read_csv("results/tsp_rl_ablation_DQN_per_instance_20250729_001452/experiment_data_20250729_001452.csv")
```

```python
# df1 = df.query("done == 1 and train_test == 'train'").copy()  # 注意：如果 train_test 是字符串，需要加引号
df1 = df.query("train_test == 'train'").copy()  # 注意：如果 train_test 是字符串，需要加引号
test_state =[
    'full', 
             'ablation_remove_current', 'ablation_remove_visited', 'ablation_remove_order', 'ablation_remove_distances', 'ablation_remove_current_visited', 'ablation_remove_current_order', 'ablation_remove_current_distances', 'ablation_remove_visited_order', 'ablation_remove_visited_distances', 'ablation_remove_order_distances']

df1['state_type'].unique()

```

```python
import numpy as np
np.set_printoptions(linewidth=np.inf, precision=4)  # 全局设置

# 任务1：按 state_type 和 instance_id 分组分析 total_reward
tmp = df1[df1['state_type'].isin(test_state)]
for state in tmp['state_type'].unique():  # 遍历所有 state_type
    
    
    # 筛选当前 state_type 的数据
    state_data = tmp[tmp['state_type'] == state]
    
    for instance in state_data['instance_id'].unique():  # 遍历该 state 下的所有 instance_id
        # 筛选当前 instance_id 的数据
        instance_data = state_data[state_data['instance_id'] == instance]
        
        # 提取 total_reward 并计算统计量
        rewards = instance_data['total_reward'].values  # 转为 numpy 数组
        if len(rewards) > 0:  # 确保有数据
            print(f"\n{state} InstanceID:{instance} {rewards[:10]}  Mean:{np.mean(rewards):.2f} Min-Max:({np.min(rewards):.2f},{np.max(rewards):.2f}) Std: {np.std(rewards):.2f}")
        else:
            print(f"\nInstance ID: {instance} (No data)")
```

```python
instance_metrics = df1.groupby(['algorithm', 'city_num', 'mode', 'state_type', 'train_test', 'instance_id']).agg({
            'total_reward': ['mean', 'std', 'max', 'min', 'count',list],
        })
instance_metrics = instance_metrics.reset_index()

tmp = instance_metrics[instance_metrics['state_type'].isin(test_state)]
tmp.head(100)
```
